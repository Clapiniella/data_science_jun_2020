```python
def check_if_done(user, kata_id):
    name_kata = None
    time = False
    yas = gd.get_completed_by_user(user)
    for i in range(len(yas['data'])):
        name_kata = gd.get_name_kata(kata_id)['name']
        if yas['data'][i]['id'] == kata_id:
            time = yas['data'][i]['completedAt']
            print(time)
            print(f"{user} has completed the kata {name_kata} at {time}")
            return time[:10]
    else:
        return False
```

```python
#df[user] = df.reset_index(level = 1)["id_kata"].apply(lambda x: check_if_done(user,x))
```

```python
def tru_fal(df,user):
    df[user] = df["id_kata"].apply(lambda x: check_if_done(user,x))
    return df
```
