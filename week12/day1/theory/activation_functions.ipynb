### ReLu

```python
def ReLu(x):
    return (abs(x) + x) / 2
```

### Sigmoid

```python
import math

def sigmoid(x):
  return 1 / (1 + math.exp(-x))
```

## TESTING

```python
import numpy as np
import matplotlib.pylab as plt

x_ = np.arange(40)
y_ = np.random.random(40) - 5
y_
```

```python
plt.plot(x_, y_)
```

```python
plt.plot(x_, ReLU(y_))
```

```python
y_sigmoid_ = [sigmoid(x) for x in y_]

plt.plot(x_, y_sigmoid_)
```
