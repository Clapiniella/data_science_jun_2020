```python
from functools import wraps
import pandas as pd

def prepost(*arg, **kwargs):
    def inner(function):
        @wraps(function)
        def wrapper(*a, **k):
            print("a", a)
            print("k", k)
            # antes
            if "url" in kwargs:
                df = pd.read_csv(kwargs["url"])
            retval = function(*a, **k)
            # después
            if "url" in kwargs:
                df.hist()
            return retval
        return wrapper
    return inner

key_url = "http://winterolympicsmedals.com/medals.csv"

@prepost(url=key_url)
def _f_protected():
    f = lambda x: x>5
    l1 = [x for x in range(16)]
    return list(filter(f, l1))

filter_ = _f_protected()
filter_
```
