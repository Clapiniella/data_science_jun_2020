### 1. Se pide:

A partir del archivo "USA_Housing.csv", realizar la regresión no lineal entre cada columna numérica con la columna 'price' utilizando:

- Regresión polinómica no lineal con grados del 2 al 10. ¿El score se estanca a partir de cierto grado?
- SVR

```python
import math
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
# SVM para problemas de regresión (SVR)
from sklearn.svm import SVR
# SVM para problemas de clasificación (SVC)
from sklearn.svm import SVC

from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler

from sklearn.metrics import r2_score, mean_absolute_error, accuracy_score
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Gabvaztor/data_science_jun_2020/master/week9/day1/data/USA_Housing.csv")
df
```

**Regresión polinómica no lineal con grados del 2 al 10. ¿El score se estanca a partir de cierto grado?**

```python
degree_min = 2
degree_max = 10

#lista de columnas numéricas, sin Price
col_num = df._get_numeric_data().columns[:-1]

dictionary_scores = {"column":[], "degree":[], "score":[]}

for col in col_num:
    for degree in range(degree_min,degree_max+1):
        print("################")
        X = np.array(df[col])
        y = df.Price

        X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2)

        polinominal_model = PolynomialFeatures(degree) 
        X_poly = polinominal_model.fit_transform(X_train.reshape(-1, 1), y_train)

        #######
        print(X_poly.shape)

        model = LinearRegression()
        model.fit(X_poly, y_train)
        y_pred = model.predict(X_poly)

        score = r2_score(y_train, y_pred)

        print("y_pred", y_pred.shape)
        print("score:", score)

        #código para crear el dataframe de score            
        lista_val = [col, degree, score]
        for i,e in enumerate(["column", "degree", "score"]):
            dictionary_scores[e].append(lista_val[i])

        X_train_to_show, y_train_to_show = zip(*sorted(zip(X_train, y_train)))
        plt.scatter(X_train_to_show, y_train_to_show, color='violet')
        X_train_to_show, y_pred = zip(*sorted(zip(X_train, y_pred)))
        plt.plot(X_train_to_show, y_pred, color='purple')
        plt.title("Poninominal degree " + str(degree))
        plt.show()
        input("Enter to next grade")
```

```python
df_score = pd.DataFrame(dictionary_scores)
df_score
```

```python
sns.lineplot(x='degree', y='score', hue='column', data=df_score)
```

Según este gráfico no se observa que se estanque notablemente a partir de cierto grado, aunque tampoco varía mucho

**SVR**

Os dejo por aquí un enlace que os puede servir: https://medium.com/pursuitnotes/support-vector-regression-in-6-steps-with-python-c4569acd062d

```python
# en este caso solo damos opciones del parámetro kernel, aunque tenemos C, gamma, degree, epsilon, coef0, entre otros 
kernel = ['rbf', 'linear', 'poly']

dicti_svr_scores = {"column":[], "kernel":[], "score":[]}

for col in col_num:
    for ker in kernel:
        print("################")
        X = np.array(df[col])
        y = np.array(df.Price)

        sc_X = StandardScaler()
        sc_y = StandardScaler()
        X = sc_X.fit_transform(X.reshape(-1, 1))
        y = sc_y.fit_transform(y.reshape(-1, 1))

        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

        #######
        print(X_train.shape)
        y_train = y_train.ravel()
        print(y_train.shape)

        model = SVR(kernel=ker)
        model.fit(X_train, y_train)
        y_pred = model.predict(X_train)

        score = r2_score(y_train, y_pred)

        print("y_pred", y_pred.shape)
        print("score:", score)

        #código para crear el dataframe de score            
        lista_val = [col, ker, score]
        for i,e in enumerate(["column", "kernel", "score"]):
            dictionary_scores[e].append(lista_val[i])

        X_train_to_show, y_train_to_show = zip(*sorted(zip(X_train, y_train)))
        plt.scatter(X_train_to_show, y_train_to_show, color='violet')
        X_train_to_show, y_pred = zip(*sorted(zip(X_train, y_pred)))
        plt.plot(X_train_to_show, y_pred, color='purple')
        plt.title("SVR with kernel: " + str(ker))
        plt.show()
        input("Enter to next kernel")
```

**3. Usando las demás columnas numéricas y una a una, utiliza SVM para clasificar esa columna modificada.**

```python
col_num_class = df_class._get_numeric_data().columns
col_num_class = col_num_class[col_num_class != 'Avg. Area Number of Bedrooms']
col_num_class
```

```python
dicti_svc_scores = {"column":[], "score":[]}

for col in col_num_class:
    print("################")
    X_svc = np.array(df_class[col]).reshape(-1,1)
    y_svc = np.array(df_class[["Avg. Area Number of Bedrooms"]])
    # para problemas de clasificación no se puede standarizar el target, pues se convierte en valores continuos

    X_train_svc, X_test_svc, y_train_svc, y_test_svc = train_test_split(X_svc, y_svc, test_size=0.2)
    
    #######
    print(X_train_svc.shape)
    y_train_svc = y_train_svc.ravel()
    print(y_train_svc.shape)

        
    # hacemos la clasificación con estos valores, en el siguiente apartado probamos más parámetros con GridsearchCV
    model_svc = SVC(kernel= "rbf", C = 1000, gamma = 0.8) 
    model_svc.fit(X_train_svc, y_train_svc)
    y_pred_svc = model_svc.predict(X_test_svc)
    
    score_svc = accuracy_score(y_test_svc, y_pred_svc)

    print("y_pred", y_pred_svc.shape)
    print("score:", score_svc)

    #código para crear el dataframe de score            
    lista_val_svc = [col, score_svc]
    for i,e in enumerate(["column", "score"]):
        dicti_svc_scores[e].append(lista_val_svc[i])

    X_train_to_show_svc, y_train_to_show_svc = zip(*sorted(zip(X_train_svc, y_train_svc)))
    plt.scatter(X_train_to_show_svc, y_train_to_show_svc, color='violet')
    X_train_to_show_svc, y_pred_svc = zip(*sorted(zip(X_train_svc, y_pred_svc)))
    plt.scatter(X_train_to_show_svc, y_pred_svc, color='purple')
    plt.title(str(col))
    plt.show()
```
